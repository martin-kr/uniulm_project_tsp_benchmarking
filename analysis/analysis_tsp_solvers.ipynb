```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
# Define the column names
column_names = ['problem', 'status', 'cputime', 'walltime', 'memory']
df_small_solver_nn = pd.read_csv('/Users/martin/uniUlm/3.Semester/themen_software_engineering/hausarbeit/uniulm_project_tsp_benchmarking/benchexec/01_small/results/benchmark_solver_nn.2025-02-19_10-40-53.results.csv', sep='\t', skiprows=2, names=column_names)
df_small_solver_nn = df_small_solver_nn.drop(0)

df_small_solver_2opt = pd.read_csv('/Users/martin/uniUlm/3.Semester/themen_software_engineering/hausarbeit/uniulm_project_tsp_benchmarking/benchexec/01_small/results/benchmark_solver_2opt.2025-02-19_10-55-52.results.csv', sep='\t', skiprows=2, names=column_names)
df_small_solver_2opt = df_small_solver_2opt.drop(0)

df_small_solver_christ = pd.read_csv('/Users/martin/uniUlm/3.Semester/themen_software_engineering/hausarbeit/uniulm_project_tsp_benchmarking/benchexec/01_small/results/benchmark_solver_christofides.2025-02-19_10-54-26.results.csv', sep='\t', skiprows=2, names=column_names)
df_small_solver_christ = df_small_solver_christ.drop(0)

df_small_solver_2approx = pd.read_csv('/Users/martin/uniUlm/3.Semester/themen_software_engineering/hausarbeit/uniulm_project_tsp_benchmarking/benchexec/01_small/results/benchmark_solver_2appox.2025-02-19_10-48-21.results.csv', sep='\t', skiprows=2, names=column_names)
df_small_solver_2approx = df_small_solver_2approx.drop(0)
```

```python
small_solver_instances = df_small_solver_nn.problem.array

def extract_number(problem_name):
    match = re.search(r'\d+', problem_name)
    return int(match.group()) if match else None

small_solver_instances_num = df_small_solver_nn.problem.apply(extract_number).to_numpy()

# nn
small_solver_nn_runtime_cpu_string = df_small_solver_nn.cputime.array
small_solver_nn_runtime_cpu = pd.to_numeric(small_solver_nn_runtime_cpu_string)

small_solver_nn_runtime_wall_string = df_small_solver_nn.walltime.array
small_solver_nn_runtime_wall = pd.to_numeric(small_solver_nn_runtime_wall_string)

small_solver_nn_memory_string = df_small_solver_nn.memory.array
small_solver_nn_memory = pd.to_numeric(small_solver_nn_memory_string)


# 2-opt
small_solver_2opt_runtime_cpu_string = df_small_solver_2opt.cputime.array
small_solver_2opt_runtime_cpu = pd.to_numeric(small_solver_2opt_runtime_cpu_string)

small_solver_2opt_runtime_wall_string = df_small_solver_2opt.walltime.array
small_solver_2opt_runtime_wall = pd.to_numeric(small_solver_2opt_runtime_wall_string)

small_solver_2opt_memory_string = df_small_solver_2opt.memory.array
small_solver_2opt_memory = pd.to_numeric(small_solver_2opt_memory_string)


# 2-approx
small_solver_2approx_runtime_cpu_string = df_small_solver_2approx.cputime.array
small_solver_2approx_runtime_cpu = pd.to_numeric(small_solver_2approx_runtime_cpu_string)

small_solver_2approx_runtime_wall_string = df_small_solver_2approx.walltime.array
small_solver_2approx_runtime_wall = pd.to_numeric(small_solver_2approx_runtime_wall_string)

small_solver_2approx_memory_string = df_small_solver_2approx.memory.array
small_solver_2approx_memory = pd.to_numeric(small_solver_2approx_memory_string)


# christ
small_solver_christ_runtime_cpu_string = df_small_solver_christ.cputime.array
small_solver_christ_runtime_cpu = pd.to_numeric(small_solver_christ_runtime_cpu_string)

small_solver_christ_runtime_wall_string = df_small_solver_christ.walltime.array
small_solver_christ_runtime_wall = pd.to_numeric(small_solver_christ_runtime_wall_string)

small_solver_christ_memory_string = df_small_solver_christ.memory.array
small_solver_christ_memory = pd.to_numeric(small_solver_christ_memory_string)
```

```python
# Scatter plot for better visualization
plt.figure(figsize=(12, 6))
plt.scatter(small_solver_instances_num, small_solver_nn_runtime_cpu, color='k', label='NN', zorder=3, marker='^')
plt.scatter(small_solver_instances_num, small_solver_2opt_runtime_cpu, color='b', label='2-Opt', zorder=3, marker='^')
plt.scatter(small_solver_instances_num, small_solver_2approx_runtime_cpu, color='r', label='2-Approx', zorder=3, marker='^')
plt.scatter(small_solver_instances_num, small_solver_christ_runtime_cpu, color='g', label='2-Approx', zorder=3, marker='^')

plt.plot(small_solver_instances_num, small_solver_nn_runtime_cpu, linestyle='-', color='k', alpha=0.5, zorder=2)
plt.plot(small_solver_instances_num, small_solver_2opt_runtime_cpu, linestyle='-', color='b', alpha=0.5, zorder=2)
plt.plot(small_solver_instances_num, small_solver_2approx_runtime_cpu, linestyle='-', color='r', alpha=0.5, zorder=2)
plt.plot(small_solver_instances_num, small_solver_christ_runtime_cpu, linestyle='-', color='g', alpha=0.5, zorder=2)

plt.xlabel('Problem')
plt.ylabel('CPU Time (s)')
plt.title('CPU Time for Different Problems')

# Adjust x-axis labels
plt.xticks(rotation=90)

plt.grid(True, linestyle='--', linewidth=0.5, zorder=1)
plt.legend()
plt.show()
```
