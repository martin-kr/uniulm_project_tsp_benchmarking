```python
def permute(arr, l=0):
    if l == len(arr) - 1:
        print(arr)
        return
    for i in range(l, len(arr)):
        arr[l], arr[i] = arr[i], arr[l]  # Swap
        permute(arr, l + 1)
        arr[l], arr[i] = arr[i], arr[l]  # Backtrack

arr = [1, 2, 3]
permute(arr)
```

```python
def next_permutation(arr):
    n = len(arr)
    i = n - 2
    while i >= 0 and arr[i] >= arr[i + 1]:
        i -= 1
    if i == -1:
        return False  # No more permutations
    j = n - 1
    while arr[j] <= arr[i]:
        j -= 1
    arr[i], arr[j] = arr[j], arr[i]  # Swap
    arr[i + 1:] = reversed(arr[i + 1:])  # Reverse suffix
    return True

arr = [1, 2, 3]
arr.sort()  # Start with the smallest permutation
while True:
    print(arr)
    if not next_permutation(arr):
        break
```

```python
def print_permutation(perm):
    print(" ".join(map(str, perm)))

def generate_next_permutation(perm, directions):
    n = len(perm)

    # Step 1: Find the largest mobile element
    largest_mobile = -1
    largest_mobile_index = -1
    for i in range(n):
        if (directions[i] == -1 and i > 0 and perm[i] > perm[i - 1]) or \
           (directions[i] == 1 and i < n - 1 and perm[i] > perm[i + 1]):
            if perm[i] > largest_mobile:
                largest_mobile = perm[i]
                largest_mobile_index = i

    # Step 2: If no mobile element found, return False (all permutations generated)
    if largest_mobile == -1:
        return False

    # Step 3: Swap the largest mobile element with the adjacent one in its direction
    swap_index = largest_mobile_index + directions[largest_mobile_index]
    perm[largest_mobile_index], perm[swap_index] = perm[swap_index], perm[largest_mobile_index]
    directions[largest_mobile_index], directions[swap_index] = directions[swap_index], directions[largest_mobile_index]

    # Step 4: Change directions of all elements larger than the largest mobile element
    for i in range(n):
        if perm[i] > largest_mobile:
            directions[i] = -directions[i]

    return True

def sjt_permutations(n):
    # Initialize permutation and direction arrays
    perm = list(range(1, n + 1))  # [1, 2, ..., n]
    directions = [-1] * n  # All directions initially set to left (-1)

    # Print the first permutation
    print_permutation(perm)

    iterations = 1

    # Generate next permutations
    while generate_next_permutation(perm, directions):
        print_permutation(perm)
        iterations += 1
    
    print("Total iterations:", iterations)

# Generate permutations for n = 4
sjt_permutations(5)
```

```python
import sys
import itertools

class SJTTSPSolver:
    def __init__(self, filename):
        distance_matrix = []
        
        with open(filename, 'r') as f:
            lines = f.readlines()
            read_distances = False  

            for line in lines: 
                line = line.strip()
                line = line.replace(':', ' ')

                if line.startswith('EOF'):
                    break
                elif line.startswith('EDGE_WEIGHT_SECTION'):
                    read_distances = True
                elif read_distances:
                    elements = line.split()
                    distance_matrix.append([int(e) for e in elements])

        self.graph = distance_matrix
        self.n = len(distance_matrix)

    def calculate_cost(self, path):
        total_cost = 0
        for i in range(self.n - 1):
            total_cost += self.graph[path[i]][path[i + 1]]
        total_cost += self.graph[path[-1]][path[0]]  # Return to start
        return total_cost

    def brute_force_tsp(self):
        min_cost = sys.maxsize
        best_path = None
        nodes = list(range(self.n))
        
        for perm in itertools.permutations(nodes):
            cost = self.calculate_cost(perm)
            if cost < min_cost:
                min_cost = cost
                best_path = perm

        return min_cost, best_path
```

```python
# Example usage
filename = "../tsplib_converted_instances/01_small/gr17_converted.tsp"

sjt_solver = SJTTSPSolver(filename)
sjt_cost, sjt_path = sjt_solver.brute_force_tsp()

print("\nBrute Force (SJT Algorithm) - Minimum Cost:", sjt_cost)
print("\nPath: ", sjt_path)
```
